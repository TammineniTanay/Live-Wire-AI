```python
import yfinance as yf
import pandas as pd

# Reload historical data for AAPL to ensure 'data' is populated
# The start and end dates are derived from the context of Step 1: Data Exploration in the notebook
data = yf.download("AAPL", start="2020-01-01", end="2023-01-01")

# Flatten multi-index columns for easier feature engineering
# yfinance sometimes returns MultiIndex columns, which can cause issues with new column assignment.
# Example: ('Close', 'AAPL') might become 'Close_AAPL'
if isinstance(data.columns, pd.MultiIndex):
    data.columns = ['_'.join(col).strip() for col in data.columns.values]

# Determine the correct 'Close' column name after potential flattening
# Default to 'Close' if 'Close_AAPL' doesn't exist (e.g., if yfinance didn't return a MultiIndex)
close_col = 'Close_AAPL' if 'Close_AAPL' in data.columns else 'Close'

# Ensure the 'Close' column exists before proceeding
if close_col not in data.columns:
    raise ValueError(f"'{close_col}' column not found in data. Available columns: {data.columns.tolist()}")

# Calculate Simple Moving Averages (SMA)
data['MA_5'] = data[close_col].rolling(window=5).mean()
data['MA_10'] = data[close_col].rolling(window=10).mean()
data['MA_50'] = data[close_col].rolling(window=50).mean()

# Calculate Daily Return percentage
data['Daily_Return'] = data[close_col].pct_change() * 100

# Drop any NaN values that result from the rolling window calculations or pct_change
data.dropna(inplace=True)

# Display the tail of the DataFrame with the new features
print(data.tail())
```

## Create Target Column

Add a 'Target' column to the DataFrame, where 1 indicates tomorrow's closing price is higher than today's, and 0 otherwise.


**Reasoning**:
To create the 'Target' column as specified, I will calculate the daily price change using the 'Close_AAPL' column and then apply a conditional logic to assign 1 if the price increases tomorrow, and 0 otherwise. Finally, I will drop any resulting NaN values.



```python
data['price_change'] = data[close_col].shift(-1) - data[close_col]
data['Target'] = (data['price_change'] > 0).astype(int)

# Drop the temporary 'price_change' column if not needed
data.drop(columns=['price_change'], inplace=True)

# Drop any NaN values that might have resulted from the shift
data.dropna(inplace=True)

print(data.head())
```

## Download new AAPL data

### Subtask:
Download Apple stock price data from 2023-01-01 to 2024-01-01 using yfinance.


**Reasoning**:
The subtask requires downloading new AAPL stock data. I will use a code block to import the yfinance library, download the data for the specified date range, and store it in a new DataFrame called `new_data`.



```python
import yfinance as yf

# Download historical data for AAPL from 2023-01-01 to 2024-01-01
new_data = yf.download("AAPL", start="2023-01-01", end="2024-01-01")

print("New AAPL data downloaded successfully.")
print(new_data.head())
```

**Reasoning**:
The next step in the overall task is to apply feature engineering (moving averages, daily return) and create the 'Target' column for the newly downloaded `new_data` DataFrame, similar to how it was done for the original `data` DataFrame. I will also handle multi-index columns if they exist.



```python
import pandas as pd

# Flatten multi-index columns for easier feature engineering if necessary
if isinstance(new_data.columns, pd.MultiIndex):
    new_data.columns = ['_'.join(col).strip() for col in new_data.columns.values]

# Determine the correct 'Close' column name after potential flattening
# Default to 'Close' if 'Close_AAPL' doesn't exist
close_col_new_data = 'Close_AAPL' if 'Close_AAPL' in new_data.columns else 'Close'

# Ensure the 'Close' column exists before proceeding
if close_col_new_data not in new_data.columns:
    raise ValueError(f"'{close_col_new_data}' column not found in new_data. Available columns: {new_data.columns.tolist()}")

# Calculate Simple Moving Averages (SMA)
new_data['MA_5'] = new_data[close_col_new_data].rolling(window=5).mean()
new_data['MA_10'] = new_data[close_col_new_data].rolling(window=10).mean()
new_data['MA_50'] = new_data[close_col_new_data].rolling(window=50).mean()

# Calculate Daily Return percentage
new_data['Daily_Return'] = new_data[close_col_new_data].pct_change() * 100

# Create the 'Target' column for new_data
# 1 if tomorrow's closing price is higher than today's, 0 otherwise
new_data['price_change'] = new_data[close_col_new_data].shift(-1) - new_data[close_col_new_data]
new_data['Target'] = (new_data['price_change'] > 0).astype(int)

# Drop the temporary 'price_change' column and any NaN values resulting from calculations
new_data.drop(columns=['price_change'], inplace=True)
new_data.dropna(inplace=True)

print("Feature engineering and Target column creation for new_data completed.")
print(new_data.head())
```

**Reasoning**:
Now that the `new_data` has been prepared with the necessary features and target variable, the next step is to load the trained Random Forest Classifier, prepare the `new_data` for prediction, make predictions, and evaluate the model's performance by calculating and printing the accuracy and classification report. I'll need to ensure the `RandomForestClassifier` is available and that the `rf_model` has been previously trained and is in the kernel state. Since `sklearn` wasn't imported in the last cell, I will import it here to ensure the necessary classes are available.



```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Define features for the new data
feature_columns = ['MA_5', 'MA_10', 'MA_50', 'Daily_Return']
X_new = new_data[feature_columns]
y_new = new_data['Target']

# Check if rf_model exists. If not, train a placeholder model.
# In a real scenario, this model would be trained in a previous step (Step 3.2).
if 'rf_model' not in locals():
    print("Random Forest model not found in current session. Training a new one for demonstration.")
    # Assuming X and y from previous data exist
    X = data[feature_columns]
    y = data['Target']
    rf_model = RandomForestClassifier(random_state=42)
    rf_model.fit(X, y)
    print("Placeholder Random Forest model trained.")

# Make predictions on the new data
y_pred_new = rf_model.predict(X_new)

# Evaluate the model on the new data
accuracy_new = accuracy_score(y_new, y_pred_new)
report_new = classification_report(y_new, y_pred_new)

print(f"\nModel performance on new data (2023-01-01 to 2024-01-01):\n")
print(f"Accuracy: {accuracy_new:.2f}")
print(f"Classification Report:\n{report_new}")
```
